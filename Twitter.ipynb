```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('twitter_training.csv')
```

```python
df=df.drop(columns=df.columns[:2])
```

```python
y=df.iloc[:,0].tolist()
x=df.iloc[:,1].tolist()
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
vec=CountVectorizer()
```

```python
le=LabelEncoder()
```

```python
x_cleaned = [item for item in x if item is not None and item is not np.nan]
```

```python
vec.fit_transform(x_cleaned)
```

```python
le.fit_transform(y)
```

```python
import pandas as pd
df = pd.DataFrame({'text': x, 'label': y})


df = df.dropna()
x_cleaned = df['text'].tolist()
y_cleaned = df['label'].tolist()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_cleaned,y_cleaned,test_size=0.2,random_state=42)
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(vec.fit_transform(x_train),le.fit_transform(y_train))
```

```python
y_pred=model.predict(vec.transform(x_test))
```

```python
acc=accuracy_score(le.transform(y_test),y_pred)
```
